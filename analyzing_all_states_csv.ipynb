```python
#@title ### Does an entire column have data missing?
missing_columns = df.columns[df.isna().all()]

print("Columns that have missing data are: ")
for col in missing_columns:
    print(col)
```

```python
#@title ### Does a certain state have data missing for an entire column?

groups = df.groupby('proj_st')
filtered_df = pd.DataFrame()

column_names = df.columns
mapping_group_empty_columns = {}

# Loop through each group and filter rows where any column is NaN
for name, group in groups:
  empty_columns = []

  for column in column_names:
    if group[column].isna().all():
      empty_columns.append(column)

  mapping_group_empty_columns[name] = empty_columns

# Display the filtered DataFrame
print(mapping_group_empty_columns)
```
