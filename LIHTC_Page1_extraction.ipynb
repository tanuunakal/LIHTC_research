```python
# @title Validating data
json_file_path = '/content/drive/MyDrive/lihtc/project_location.json'
import random

def pick_random_entities(lst, num_entities):
    if num_entities > len(lst):
        raise ValueError("Number of entities to pick exceeds the length of the list.")
    return random.sample(lst, num_entities)

# Example usage:
with open(json_file_path, 'r') as f:
      data = json.load(f)
my_list = data
random_entities = pick_random_entities(my_list, 20)
for rando in random_entities:
  print(rando['gdrive_id'])
  print(rando['file_id'])
  if 'Site Geo Coordinates' in rando:
    print(rando['Site Geo Coordinates'])
  print()
```

```python
list_of_all_ids = [item['file_id'] for item in my_list]
len(list_of_all_ids)
```

```python
def pick_unique_random_entities(list_of_all_ids, num_entities):
    available_ids = [id for id in list_of_all_ids if id not in list_of_ids]

    if len(available_ids) < num_entities:
        raise ValueError("Not enough unique entities available.")

    return random.sample(available_ids, num_entities)

validation_pass_2_ids = pick_unique_random_entities(list_of_all_ids, 20)
```
